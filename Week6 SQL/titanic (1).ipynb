```python
fig, ax = plt.subplots()

ax.hist(titanic_df['Age'], bins=40)
ax.set_xlabel('Ages')
```

```python
not_survived_df = titanic_df[titanic_df['Survived'] == 0]
survived_df = titanic_df[titanic_df['Survived'] == 1]
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
fig.set_size_inches(15,7)
```

```python
ax[0].hist(not_survived_df['Age'], bins=40, facecolor='g')
ax[0].set_xlabel('Ages')
ax[0].set_title('Not survived')

ax[1].hist(survived_df['Age'], bins=40)
ax[1].set_xlabel('Ages')
ax[1].set_title('Survived')
```

```python
labels = ['Survived', 'Not Survived']
```

```python
first_class = titanic_df[titanic_df['Pclass'] == 1]


firstC_survived = len(first_class[first_class['Survived'] == 1])
firstC_not_survived = len(first_class[first_class['Survived'] == 0])

data1 = [firstC_survived, firstC_not_survived]
```

```python
second_class = titanic_df[titanic_df['Pclass'] == 2]

secondC_survived = len(second_class[second_class['Survived'] == 1])
secondC_not_survived = len(second_class[second_class['Survived'] == 0])

data2 = [secondC_survived, secondC_not_survived]
```

```python
third_class = titanic_df[titanic_df['Pclass'] == 3]
labels = ['Survived', 'Not Survived']

thirdC_survived = len(third_class[third_class['Survived'] == 1])
thirdC_not_survived = len(third_class[third_class['Survived'] == 0])

data3 = [thirdC_survived, thirdC_not_survived]
```

```python
figure, ax = plt.subplots(nrows=3, ncols=1)
figure.set_size_inches(10,15)
```

```python
ax[0].pie(data1, labels = labels, autopct='%1.1f%%', colors = ['purple','grey'])
ax[0].set_title('First Class')

ax[1].pie(data2, labels = labels, autopct='%1.1f%%', colors = ['purple','grey'])
ax[1].set_title('Second Class')

ax[2].pie(data3, labels = labels, autopct='%1.1f%%', colors = ['purple','grey'])
ax[2].set_title('Third Class')
```
